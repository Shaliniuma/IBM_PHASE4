# Import Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

# Import the Data to the Notebook

```python
data=pd.read_csv("CAD_Phase4.csv")
data.head()
```

```python
feed_data=data.drop('Price',axis=1)
feed_data.head()
```

```python
output=data.Price
output.head()
```

# Visualizations of the Data

```python
plt.scatter(data['AvgIncome'],data['Price'])
plt.xlabel("Average Income of the area")
plt.ylabel("Price")
plt.title("AvgIncome VS Price");
```

```python
plt.scatter(data['AvgHouse Age'],data['Price'])
plt.xlabel("Average Age of the House")
plt.ylabel("Price")
plt.title("AvgHouseAge VS Price");
```

```python
plt.scatter(data['AvgRooms'],data['Price'])
plt.xlabel("Average Number of Rooms")
plt.ylabel("Price")
plt.title("AvgRooms VS Price");
```

```python
plt.scatter(data['AvgBedrooms'],data['Price'])
plt.xlabel("Average Number of the Bedrooms")
plt.ylabel("Price")
plt.title("AvgBedrooms VS Price");
```

```python
plt.scatter(data['AreaPopulation'],data['Price'])
plt.xlabel("Average Population of the area")
plt.ylabel("Price")
plt.title("AreaPopulation VS Price");
```

# Creating the Predection Model

```python
reg=linear_model.LinearRegression()
reg.fit(data[['AvgIncome','AvgHouse Age','AvgRooms','AvgBedrooms','AreaPopulation']],data.Price)
```

# Predection

```python
a=int(input("Enter the average income of the area:"))
b=int(input("Enter the average House Age:"))
c=int(input("Enter the Number of Rooms avilable:"))
d=int(input("Enter the average Number of Bedrooms:"))
e=int(input("Enter the Population of the area:"))
print("The Average Rate of the House with these specifications would be",reg.predict([[a,b,c,d,e]]),"$")
```

# Visualizations of the Predection

```python
predection=reg.predict(data[['AvgIncome','AvgHouse Age','AvgRooms','AvgBedrooms','AreaPopulation']])
plt.scatter(data.Price,predection)
plt.xlabel("Original Price")
plt.ylabel("Predected Price")
plt.title("Original Price Vs Predected Price");
```
